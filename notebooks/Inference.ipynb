```python
import boto3
import sagemaker
from sagemaker.predictor import Predictor
from sagemaker.serializers import JSONSerializer
from sagemaker.deserializers import JSONDeserializer

region = "us-east-1"
endpoint_name = "xgboost-excalibur"

sagemaker_session = sagemaker.Session()
sagemaker_client = boto3.client("sagemaker", region_name=region)

# Verificar que existe
existing_endpoints = sagemaker_client.list_endpoints(NameContains=endpoint_name)["Endpoints"]
endpoint_names = [ep["EndpointName"] for ep in existing_endpoints]

if endpoint_name not in endpoint_names:
    raise ValueError(f"Endpoint '{endpoint_name}' no existe.")

print(f"Conectado al endpoint: {endpoint_name}")

predictor = Predictor(
    endpoint_name=endpoint_name,
    sagemaker_session=sagemaker_session,
    serializer=JSONSerializer(),
    deserializer=JSONDeserializer()
)
```

Test payload

```python
payload_json = {
    "did_you_get_injured_byaslip_or_fall_accident": None,
    "did_you_have_an_accident_at_work": False,
    "how_you_were_involved": None,
    "days_since_accident": 201,
    "state_accident_occur": "New York",
    "were_you_affected_by_possible_malpractice": True,
    "were_you_involved_in_an_automobile_accident": None
}
```

Predict

```python
response = predictor.predict(payload_json)

print("Prediction:")
print(response)
```
